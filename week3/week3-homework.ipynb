```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('responses.csv', delimiter = ",") 
```

```python
df
```

```python
df.dtypes
```

```python
df['min'] = pd.to_numeric(df['min'], errors = 'coerce') # change min from string to numeric
df
```

```python
# create a column "ttime" and add minutes + seconds
df['ttime'] = df['min']*60 + df.sec # make a ttime variable and ttime= min + sec
df
```

```python
# This first row in the DataFrame has a 68 seconds which is wrong. Need to convert it into minutes. 
# If we divide it with 60, add 1 to minues and the remaining 8 to seconds 
# calculate minutes (in other words Quotient)
df['min'] = df['ttime'] // 60 # (// Floor division) => 248/60 = 4

# calculate seconds (in other words Remainder)
df['sec'] = df['ttime'] % 60 # (% Modulus) => 248/60 = 8
```

```python
df
```

```python
df.drop(['ttime'], axis=1) # ttime is not required anymore so dropping
```

```python
# In rirst variables "exp" and "cor" has both yes & now values. It should have only one of these.
# therefore replacing yes/now with NaN
df.loc[~df['exp'].isin(['Yes','No']),'exp'] = np.nan ## if exp has yes,no values replace with NaN
```

```python
df
```

```python
# variable "cor" shall have only yes or no. Not both. Therefor replacing with NaN
df.loc[~df['cor'].isin(['Yes','No']),'cor'] = np.nan ## if cor has yes,no values replace with NaN
```

```python
df
```

```python
df.drop([14]) # removing row 14 because it has all NAN values. 
```

```python
# Cleaning dataset. Removing all those that don't have time. 
not_null_df = df[df['ttime'].notnull()] # extract only those that have time
```

```python
# combine all rows entries for Green sudoku
df_greek=not_null_df.loc[not_null_df.type == 'Greek']
```

```python
# combine all rows entries for Latin sudoku
df_latin=not_null_df.loc[not_null_df.type == 'Latin']
```

```python
mean_greek = df_greek['ttime'].mean(axis = 0, skipna = True) # calculate mean of Greek players
mean_latin = df_latin['ttime'].mean(axis = 0, skipna = True) # calculate mean of Latin players
mean_greek, mean_latin
```

```python
# Find the faster Sudoku and prints
faster = "Greek are faster" if mean_greek > mean_latin else "Latin are faster"
print(faster)
```
